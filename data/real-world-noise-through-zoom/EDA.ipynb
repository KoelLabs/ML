```python
import glob
```

```python
# all .txt files with only one . in the name
for name in glob.glob('*.c0.txt'):
    for ext in ['words.txt', 'c0.txt', 'c1.txt', 'taguchi.txt', 'facebook60.txt', 'rubino.txt', 'vitou.txt']:
        with open(name.replace('c0.txt', ext)) as f:
            transcription = f.read().replace('\n', '')
            if ext == 'words.txt':
                print(transcription)
                print('---')
            else:
                print(transcription.replace(' ', '') + '\t\t' + ext.replace('.txt', ''))
    print()
```
