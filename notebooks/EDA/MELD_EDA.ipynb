```python
import tarfile
import pandas as pd
```

```python
DATA_TAR = '../../.data/MELD.Raw.tar.gz'
```

```python
# print the contents of the tar file
with tarfile.open(DATA_TAR, 'r') as tar:
    tar.list()
```

```python
# print contents of README.txt
with tarfile.open(DATA_TAR, 'r') as tar:
    readme = tar.extractfile('MELD.Raw/README.txt')
    print(readme.read().decode('utf-8'))
    readme.close()
```

```python
# load the test_sent_emo.csv file into a pandas dataframe
with tarfile.open(DATA_TAR, 'r') as tar:
    train_sent_emo = tar.extractfile('MELD.Raw/test_sent_emo.csv')
    df = pd.read_csv(train_sent_emo)
    train_sent_emo.close()
df.head()
```

```python
# print content of test.tar.gz
with tarfile.open(DATA_TAR, 'r') as tar:
    inner_tar = tar.extractfile('MELD.Raw/test.tar.gz')
    with tarfile.open(fileobj=inner_tar, mode='r') as inner:
        inner.list()
    inner_tar.close()
```

```python
DIALOG_ID = 2
```

```python
dialog_df = df[df['Dialogue_ID'] == DIALOG_ID]
dialog_df.to_csv(f'../../data/dialog_{DIALOG_ID}.csv', index=False, sep=';')
dialog_df
```

```python
# extract the utterances from the test.tar.gz file corresponding to the DIALOG_ID
with tarfile.open(DATA_TAR, 'r') as tar:
    inner_tar = tar.extractfile('MELD.Raw/test.tar.gz')
    with tarfile.open(fileobj=inner_tar, mode='r') as inner:
        for utterance_id in df[df['Dialogue_ID'] == DIALOG_ID]['Utterance_ID']:
            utterance = inner.extractfile(f'./output_repeated_splits_test/dia{DIALOG_ID}_utt{utterance_id}.mp4')
            with open(f'../../data/utterance{utterance_id}.mp4', 'wb') as f:
                f.write(utterance.read())
            utterance.close()
    inner_tar.close()
```
